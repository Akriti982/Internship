```python
# Source Code
import requests
from bs4 import BeautifulSoup
import pandas as pd
#To scrape all types of textual content
def web_scrape_func(website, html, class_):
    page=requests.get(website)
    soup=BeautifulSoup(page.content)
    lst=[]
    for x in soup.find_all(html,class_):
        lst.append(x.text)
    return page, soup, lst
#To tabulate scraped content
def create_dataframe(data, **kwargs):
    data=pd.DataFrame(data, **kwargs)
    return data
#To scrape all types of links
def links_img_video(website, html, class_, sub_html):
    page=requests.get(website)
    soup=BeautifulSoup(page.content)
    links=[]
    for x in soup.find_all(html, class_):
        links.append(x[sub_html])
    return page, soup, links
```
