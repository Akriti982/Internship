```python
description=driver.find_elements_by_xpath('//p[@class="col-9 color-fg-muted my-1 pr-4"]')
description
```

```python
repository_description=[]
for x in description:
    repository_description.append(x.text)
repository_description
```

```python
repository_description.insert(0,'_')
repository_description
```

```python
import time
from selenium.common.exceptions import NoSuchElementException
contributors=[]
for x in repository_title:
    try:
        driver.find_element_by_link_text(x).click()
        cont=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div/div[3]/div[2]/div/div[4]/div/h2/a/span')
        contributors.append(cont.text)
        driver.back()
        time.sleep(10)
    except NoSuchElementException:
        contributors.append('_')
        driver.back()
        continue
contributors
```

```python
contributors.count('')
```

```python
for x in range(len(contributors)):
    if contributors[x]=='':
        print(x)
```

```python
driver.execute_script('window.scrollBy(0,-3000)')
```

```python
counter=[3,12,18,21]
import time
from selenium.common.exceptions import NoSuchElementException
con=[]
for x in counter:
    try:
        driver.find_element_by_link_text(repository_title[x]).click()
        cont=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div/div[3]/div[2]/div/div[6]/div/h2/a/span')
        con.append(cont.text)
        driver.back()
        time.sleep(10)
    except NoSuchElementException:
        con.append('_')
        driver.back()
        continue
con
```

```python
contributors.remove('')
contributors.insert(3, con[0])
contributors
```

```python
contributors.remove('')
contributors.insert(12, con[1])
contributors
```

```python
contributors.remove('')
contributors.insert(18, con[2])
contributors
```

```python
contributors.remove('')
contributors.insert(21, con[3])
contributors
```

```python
driver.execute_script('window.scrollBy(0,-3000)')
```

```python
import time
from selenium.common.exceptions import NoSuchElementException
languages=[]
for x in repository_title:
    try:
        driver.find_element_by_link_text(x).click()
        cont=driver.find_elements_by_xpath('//li[@class="d-inline"]')
        languages.append('{}'.format(x))
        for w in cont:
            languages.append(w.text)
        driver.back()
        time.sleep(10)
    except NoSuchElementException:
        languages.append('_')
        driver.back()
        continue
languages
```

```python
lst=[]
for x in range(len(languages)):
    for w in repository_title:
        if languages[x]==w:
            lst.append(x)
print(lst)
```

```python
len(repository_title), len(lst)
```

```python
lang={}
for x in range(len(lst)):
    try:
        lang.update({languages[lst[x]]:languages[lst[x]+1:lst[x+1]]})
    except IndexError:
        continue
lang
```

```python
languages[-3:]
```

```python
language=[]
for x in repository_title:
    try:
        language.append(lang[x])
    except KeyError:
        language.append(languages[-2:])
language
```

```python
solution=pd.DataFrame(language, columns=['Language1', 'Language2', 'Language3', 'Language4', 'Language5', 'Language6', 'Language7'])
solution
```

```python
solution['Contributors Count']=contributors
solution['Repository Description']=repository_description
solution['Repository Title']=repository_title
solution=solution[['Repository Title', 'Repository Description', 'Contributors Count', 'Language1', 'Language2', 'Language3', 'Language4', 'Language5', 'Language6', 'Language7']]
solution
```

```python
driver.close()
```
