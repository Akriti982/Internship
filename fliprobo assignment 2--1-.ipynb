```python
ratings_2_10=[]
x=1
while x<11:
    rate=driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq" or @class="_3LWZlK _1rdVr6 _1BLPMq"]')
    new_rate=[]
    for w in rate:
        new_rate.append(w.text)
    ratings_2_10.append(new_rate)
    driver.refresh()
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[12]').click()
    print(x)
    x+=1
    print(ratings_2_10)
```
